# Pandas desde la Programacin Orientada a Objetos

Este notebook introduce **Pandas** desde una perspectiva de **Programacin Orientada a Objetos (POO)**, mostrando cmo sus estructuras principales encapsulan datos y comportamiento para el anlisis de datos.

## 1. Importancia de Pandas en el ecosistema cientfico

Pandas es una biblioteca clave para la manipulacin y anlisis de datos estructurados. Acta como puente entre NumPy y bibliotecas de visualizacin y Machine Learning.

```
import pandas as pd
import numpy as np

print(pd.__version__)
```

## 2. Pandas y la POO

En Pandas, los datos se representan mediante **objetos especializados**:

- `Series`: estructura unidimensional etiquetada.
- `DataFrame`: estructura bidimensional con filas e ndices.

Ambos encapsulan datos, ndices, tipos y mtodos.

```
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s
```

```
df = pd.DataFrame({
    'nombre': ['Ana', 'Luis', 'Mara'],
    'edad': [23, 35, 29]
})
df
```

## 3. Encapsulacin y mtodos

Las operaciones sobre los datos se realizan mediante **mtodos del objeto**, no manipulando directamente estructuras internas.

```
df.describe()
```

## 4. Polimorfismo: operaciones coherentes

Muchos mtodos funcionan de forma similar en `Series` y `DataFrame`, lo que refleja el principio de **polimorfismo**.

```
s.mean(), df['edad'].mean()
```

## 5. Operaciones alineadas por etiquetas

Pandas alinea los datos automticamente usando ndices y columnas.

```
s2 = pd.Series([5, 5, 5], index=['b', 'c', 'd'])
s + s2
```

## 6. Indexacin explcita

`loc` y `iloc` ofrecen interfaces claras para acceder al estado interno del objeto.

```
df.loc[0, 'nombre'], df.iloc[1, 1]
```

## 7. Copias y referencias

Comprender cundo se crea una copia o una vista es esencial en el modelo de objetos de Pandas.

```
df2 = df
df2['edad'] = df2['edad'] + 1
df
```

```
df3 = df.copy()
df3['edad'] = df3['edad'] + 10
df, df3
```

## 8. Conclusin

Pandas es un ejemplo slido de **POO aplicada a Ciencia de Datos**, donde los objetos representan entidades complejas con comportamiento bien definido.
